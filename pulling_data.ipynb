# Extracting data from Alpaca

```python
import pandas as pd
from Alpaca_account import Alpaca_account
import json
from datetime import datetime
```

```python
my_account = Alpaca_account("dl.cfg")
```

```python
my_account.get_bars_data("15Min", "AAPL", "100")
```

```python
my_account.get_bars_data("15Min", "AAPL,MSFT", "100")
```

```python
json_data = json.dumps(my_account.get_bars_data("15Min", "AAPL,MSFT", "100"), indent=4)
```

```python
data_df = pd.read_json(json_data)
```

```python
data_df.head()
```

```python
 data_df["AAPL"]
```

```python
type(data_df['AAPL'])
```

```python
closing_prices = []
opening_prices = []
date_times = []

for i in data_df['AAPL']:
    closing_prices.append(i['c'])
    opening_prices.append(i['o'])
    date_times.append(datetime.fromtimestamp(i['t']))
```
